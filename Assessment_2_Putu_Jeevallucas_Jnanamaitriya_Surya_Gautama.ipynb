````python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression

# Menetapkan style Seaborn
plt.style.use("seaborn-v0_8")

# URL dari dataset
url = "https://gml.noaa.gov/webdata/ccgg/trends/co2/co2_mm_mlo.txt"

# Membaca dataset ke dalam DataFrame
df = pd.read_csv(url, delim_whitespace=True, comment='#', header=None, names=[
    "Year",
    "Month",
    "Decimal_Date",
    "Monthly_Average",
    "De_seasonalized",
    "Days",
    "Std_Dev",
    "Uncertainty_of_Mean"])

# Menampilkan 5 baris pertama dan terakhir dari DataFrame
print(df)
````

# **Part 1**: Descriptive Statistics - Masic Measures
```
Compute the mean, median, and standard deviation for CO2 Levels.
```
Kode dimulai dengan menghitung **tiga ukuran statistik penting**: *mean*, median, dan *standard deviation* dari **CO2 Levels**. Ukuran-ukuran ini memberikan **wawasan kunci** tentang distribusi dan karakteristik **dataset**.

- **Mean**: Dihitung menggunakan fungsi `mean(`) pada kolom `Monthly_Average` dari **DataFrame df**. Rata-rata ini **mewakili** rata-rata tingkat CO2 di **seluruh** titik data.

- **Median**: Dihitung menggunakan fungsi `median()`. Median mewakili **nilai tengah** dalam dataset ketika diurutkan.

- **Standard Deviation**: Dihitung menggunakan fungsi `std()`. Ini mengukur **tingkat variasi** atau **dispersi** dalam CO2 Levels. Deviasi standar yang lebih tinggi menunjukkan **variabilitas yang lebih besar**.

````python
# Menghitung mean, median, dan standard deviation CO2 Levels
co2_mean = df['Monthly_Average'].mean()
co2_median = df['Monthly_Average'].median()
co2_std = df['Monthly_Average'].std()

# Menampilkan statistik yang dihitung dengan dua tempat desimal
print(f"Mean of CO2 Levels: {co2_mean:.2f}")
print(f"Median of CO2 Levels: {co2_median:.2f}")
print(f"Standard Deviation of CO2 Levels: {co2_std:.2f}")
````

## **Tingkat CO2 Minimum, Median, dan Maksimum**

Untuk memahami variasi dalam tingkat CO2 selama beberapa tahun, pertama-tama identifikasi tahun dengan tingkat CO2 minimum, median, dan maksimum:

- **Tahun dengan Tingkat CO2 Minimum**: Dalam dataset ini, tahun dengan tingkat rata-rata CO2 bulanan terendah adalah **`{{min_year}}`**. Ini menunjukkan titik ketika konsentrasi CO2 berada pada level terendah.

- **Tahun dengan Tingkat CO2 Median**: Tahun dengan tingkat rata-rata CO2 bulanan median adalah **`{{mid_year}}`**. Tahun ini mencerminkan titik tengah konsentrasi CO2, menunjukkan kapan tingkatnya tidak terlalu rendah atau tinggi.

- **Tahun dengan Tingkat CO2 Maksimum**: Tahun dengan tingkat rata-rata CO2 bulanan tertinggi adalah **`{{max_year}}`**. Tahun ini menandai titik di mana konsentrasi CO2 mencapai puncaknya selama periode yang diamati.

## **Persentase Kenaikan Tingkat CO2**

Menghitung persentase kenaikan tingkat CO2 dari tahun dengan tingkat minimum hingga tahun dengan tingkat maksimum. Perhitungan ini memberikan perspektif yang jelas tentang pertumbuhan keseluruhan dalam konsentrasi CO2 dari waktu ke waktu:

- **Persentase Kenaikan**: Konsentrasi CO2 telah meningkat sekitar **`{{percentage_increase:.2f}}%`** dari tahun dengan tingkat minimum ke tahun dengan tingkat maksimum. Statistik ini menyoroti peningkatan yang mengkhawatirkan dalam tingkat CO2 selama beberapa tahun yang dianalisis.

````python
# Temukan tahun dengan tingkat CO2 minimum, tengah, dan maksimum
min_year = df[df['Monthly_Average'] == df['Monthly_Average'].min()]['Year'].values[0]
mid_year = df[df['Monthly_Average'] == df['Monthly_Average'].median()]['Year'].values[0]
max_year = df[df['Monthly_Average'] == df['Monthly_Average'].max()]['Year'].values[0]

# Hitung persentase peningkatan tingkat CO2 dari minimum ke maksimum
min_co2 = df['Monthly_Average'].min()
max_co2 = df['Monthly_Average'].max()
percentage_increase = ((max_co2 - min_co2) / min_co2) * 100

# Buat line plot
plt.figure(figsize=(14, 6))
sns.lineplot(data=df, x="Year", y="Monthly_Average", label='CO2 Levels')
plt.title("Relationship Between Year and CO2 Levels")
plt.xlabel("Year")
plt.ylabel("CO2 Levels")

# Anotasi tahun-tahun signifikan dengan teks
significant_years = [min_year, mid_year, max_year]
for year in significant_years:
    plt.text(year, df.loc[df['Year'] == year, 'Monthly_Average'].values[0] + 3, f'{year}', ha='center', va='bottom')

# Tambahkan label teks untuk persentase peningkatan CO2 Levels
plt.text(max_year, max_co2 - 10, f'(+{percentage_increase:.2f}%)', ha='center', va='top', color='red')

plt.show()
````

## **Insights**
- Line Plot **dengan jelas** menggambarkan **tren meningkatnya tingkat CO2** dari tahun ke tahun. Tren ini mengindikasikan **masalah berkelanjutan** dari peningkatan **konsentrasi karbon dioksida** dalam atmosfer, yang merupakan salah satu **penyebab utama perubahan iklim**.
- Analisis mengidentifikasi **tiga tahun yang signifikan** dalam hal tingkat CO2: tahun dengan tingkat **minimum**, **median**, dan **maksimum**. Tahun-tahun ini berfungsi sebagai **tonggak penting** dalam kumpulan data.
- Peningkatan persentase dalam tingkat CO2 dari tahun dengan **minimum** hingga tahun dengan **maksimum** adalah sekitar **35.71%**. Peningkatan yang **signifikan** ini menekankan **percepatan** konsentrasi CO2 dari waktu ke waktu, memperkuat **pentingnya** tindakan iklim dan strategi **mitigasi**.

## **Interpretations**
Line Plot **memberikan representasi yang jelas dan informatif** tentang bagaimana tingkat CO2 telah **berubah seiring waktu**, memperlihatkan **tren meningkat secara keseluruhan** dan **tahun-tahun kunci yang menarik**.

# **Part 4**: Supervised Learning - Regression - Simple Linear Regression

```
Perform a simple linear regression using Year as the independent variable and CO2 Levels as the dependent variable.
```

## **Membuat Scatter Plot**
Scatter Plot dibuat menggunakan Seaborn. Plot ini menampilkan hubungan antara "Year" pada sumbu-x dan tingkat CO2 "Monthly_Average" pada sumbu-y.

## **Menyesuaikan Model Linear Regression**
Model LinearRegression dari scikit-learn diinisialisasi sebagai `lm`. Variabel independen X didefinisikan sebagai nilai "Year", dan variabel dependen y diatur sebagai tingkat CO2 "Monthly_Average". Model Linear Regression kemudian disesuaikan dengan data menggunakan `lm.fit(X, y)`.

````python
# Membuat scatter plot
plt.figure(figsize=(14, 6))
sns.scatterplot(data=df, x="Year", y="Monthly_Average", label='CO2 Levels')
plt.title("Linear Regression using Year and CO2 Levels")
plt.xlabel("Year")
plt.ylabel("CO2 Levels")

# Menyesuaikan model linear regression
lm = LinearRegression()
X = df["Year"].values.reshape(-1, 1)
y = df["Monthly_Average"].values
lm.fit(X, y)

# Menampilkan linear regression
sns.lineplot(data=df, x="Year", y=lm.predict(X), color='red', label='Linear Regression')

plt.show()
````

## **Insights**
- Scatterplot menunjukkan bahwa tingkat CO2 secara konsisten meningkat seiring waktu, yang mengindikasikan masalah berkelanjutan dari emisi karbon yang berkontribusi terhadap perubahan iklim.

- Kode ini adalah alat berharga untuk memahami dan memvisualisasikan data historis tingkat CO2, yang dapat menjadi kunci dalam penelitian lingkungan dan pengambilan keputusan kebijakan.

# **Part 5**: Unsupervised Learning - K-Means Clustering

```
If there are other climate-related variables in the dataset, use K-Means clustering to categorize years into different climate profiles.
```

Menggunakan **K-Means Clustering** untuk mengkategorikan `Year` ke dalam profil iklim yang berbeda berdasarkan variabel terkait iklim lainnya dalam dataset. Sebelum menerapkan **K-Means Clustering**, perlu mempersiapkan data dengan memilih fitur-fitur yang relevan dan menyesuaikannya jika diperlukan.

Identifikasi variabel terkait iklim yang ingin digunakan untuk pengelompokan. Dapat memilih kolom seperti `"Year"`, `"Monthly_Average"`, `"De_seasonalized"`, `"Days"`, `"Std_Dev"`, dan `"Uncertainty_of_Mean"`.

````python
# Pilih kolom-kolom yang relevan untuk Clustering
data = df[["Year", "Monthly_Average", "De_seasonalized", "Days", "Std_Dev", "Uncertainty_of_Mean"]]
````

Menormalkan data karena **K-Means** sensitif terhadap skala variabel. Dapat menggunakan `StandardScaler` dari `scikit-learn` untuk menormalkan data.

````python
from sklearn.preprocessing import StandardScaler

# Standarisasi data
scaler = StandardScaler()
scaled_data = scaler.fit_transform(data)
````

Untuk menentukan jumlah **Clusters** optimal pada model **K-Means**, dapat menggunakan metode seperti metode **Elbow** atau **Silhouette Score**.

````python
from sklearn.cluster import KMeans
import numpy as np

# Temukan jumlah Cluster (K) optimal menggunakan metode Elbow
inertia = []
for k in range(1, 11):
    kmeans = KMeans(n_clusters=k, random_state=0, n_init=9)
    kmeans.fit(scaled_data)
    inertia.append(kmeans.inertia_)

# Plot grafik Elbow untuk melihat Elbow Point
plt.figure(figsize=(14, 6))
sns.lineplot(x=range(1, 11), y=inertia, marker='o', markers=True, label='Intertia')
plt.title('Elbow Method')
plt.xlabel('Number of Clusters (K)')
plt.ylabel('Inertia')
plt.legend()

plt.show()
````

Berdasarkan metode **Elbow**, pilih **Clusters (K)** yang optimal. Dalam grafik, **Elbow Point** adalah di mana inersia mulai berkurang dengan laju yang lebih lambat. Jadi, **K adalah 3** (optimal).

````python
# Temukan jumlah Cluster (K) optimal menggunakan metode Elbow
inertia = []
for k in range(1, 11):
    kmeans = KMeans(n_clusters=k, random_state=0, n_init=10)
    kmeans.fit(scaled_data)
    inertia.append(kmeans.inertia_)

# Identifikasi Elbow Point dengan menginspeksi visual plot
knee = 3  # Menyesuaikan nilai ini berdasarkan inspeksi visual

# Plot grafik Elbow
plt.figure(figsize=(14, 6))
sns.lineplot(x=range(1, 11), y=inertia, marker='o', markers=True, label='Intertia')
plt.title('Elbow Method')
plt.xlabel('Number of Clusters (K)')
plt.ylabel('Inertia')

# Meng-highlight Elbow Point
plt.axvline(x=knee, color='red', linestyle=':', label='Elbow Point', linewidth=1)
plt.legend()

plt.show()
````

Jumlah **Clusters (K)** optimal telah ditentukan, sekarang dapat melakukan **K-Means Clustering**.

````python
kmeans = KMeans(n_clusters=knee, random_state=0, n_init=10)  # Tetapkan n_init ke nilai default sebelumnya
kmeans.fit(scaled_data)

# Tambahkan label 'Cluster' ke DataFrame asli
df["Cluster"] = kmeans.labels_
````

Memvisualisasikan **Clusters (K)** dengan membuat **Scatter Plot**.

````python
# Year dengan Days
plt.figure(figsize=(14, 6))
sns.scatterplot(x="Monthly_Average", y="De_seasonalized", hue="Cluster", data=df, alpha=0.5)
plt.title("CO2 Levels and its De-seasonalized (K-Means Clustering)")
plt.xlabel("CO2 Levels")
plt.ylabel("De-seasonalized")
plt.show()
print("\n")  # Menambahkan baris kosong untuk spacing

# Year dengan Days
plt.figure(figsize=(14, 6))
sns.scatterplot(x="Year", y="Days", hue="Cluster", data=df, alpha=0.5)
plt.title("Year and Days (CO2 Levels K-Means Clustering)")
plt.xlabel("Year")
plt.ylabel("Days")
plt.show()
print("\n")  # Menambahkan baris kosong untuk spacing

# Year dengan Std_Dev
plt.figure(figsize=(14, 6))
sns.scatterplot(x="Year", y="Std_Dev", hue="Cluster", data=df, alpha=0.5)
plt.title("Year and Standard Deviation (CO2 Levels K-Means Clustering)")
plt.xlabel("Year")
plt.ylabel("Standard Deviation")
plt.show()
print("\n")

# Year dengan Uncertainty of Mean
plt.figure(figsize=(14, 6))
sns.scatterplot(x="Year", y="Uncertainty_of_Mean", hue="Cluster", data=df, alpha=0.5)
plt.title("Year and Uncertainty of Mean (CO2 Levels K-Means Clustering)")
plt.xlabel("Year")
plt.ylabel("Uncertainty of Mean")
plt.show()
````

## **Insights**
Keempat kategori **K-Means Clustering** mengungkapkan hasil penerapan pada **CO2 Levels**, dengan kelompok yang berbeda ditampilkan dalam warna yang berbeda. Hal ini memberitahu bahwa data dapat dikelompokkan menjadi
3 kelompok berdasarkan kesamaan **CO2 Levels**.

## **Interpretations**
Dengan menganalisis **Scatter Plot**, tampak jelas bahwa **K-Means** berhasil memisahkan data menjadi kelompok-kelompok yang berbeda. Ini dapat menjadi sangat berguna dalam mengidentifikasi periode-periode dengan perilaku atau tren **CO2** yang serupa.
